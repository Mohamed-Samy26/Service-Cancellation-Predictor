```python
# step 5 categoral data
#_________________________________________________________________________________
df.Contract = pd.Categorical(df.Contract , ['Month-to-month','One year','Two year'],ordered=True)
df.Contract = df.Contract.cat.codes

df.gender = pd.Categorical(df.gender , ['Male','Female'],ordered=True)
df.gender = df.gender.cat.codes

df.PaymentMethod = pd.Categorical(df.PaymentMethod , ['Electronic check','Mailed check','Bank transfer (automatic)','Credit card (automatic)'] ,ordered=True)
df.PaymentMethod = df.PaymentMethod.cat.codes

df.InternetService = pd.Categorical(df.InternetService , ['DSL','Fiber optic', 'No'] ,ordered=True)
df.InternetService = df.InternetService.cat.codes
```

```python
#Convert all columns have Yes or No to 0 and 1
df = df.replace(to_replace = ['Yes','No'],value = ['1','0'])
```

```python
df['Partner'] = df['Partner'].astype(int)
df['Dependents'] = df['Dependents'].astype(int)
df['PhoneService'] = df['PhoneService'].astype(int)
df['MultipleLines'] = df['MultipleLines'].astype(int)
df['OnlineSecurity'] = df['OnlineSecurity'].astype(int)
df['OnlineBackup'] = df['OnlineBackup'].astype(int)
df['DeviceProtection'] = df['DeviceProtection'].astype(int)
df['TechSupport'] = df['TechSupport'].astype(int)
df['StreamingTV'] = df['StreamingTV'].astype(int)
df['PaperlessBilling'] = df['PaperlessBilling'].astype(int)
df['PaperlessBilling'] = df['PaperlessBilling'].astype(int)
```

```python
df.dtypes
```

```python

# copy the data
df_z_scaled = df.copy()
columns = ["tenure" , "MonthlyCharges","TotalCharges"]
# apply normalization techniques
for column in columns:
    df_z_scaled[column] = (df_z_scaled[column] -
                           df_z_scaled[column].mean()) / df_z_scaled[column].std()
  
# view normalized data   
display(df_z_scaled)
```

```python
 df_z_scaled.to_csv('output_with_scaling.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import joblib 
x = df_z_scaled.drop('Churn', axis=1)
y = df_z_scaled['Churn']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=200)
cls = svm.SVC(kernel="linear")
cls.fit(x_train, y_train)
pred = cls.predict(x_test)
print("accuracy ::", metrics.accuracy_score(y_test, y_pred=pred))
print("precision ::", metrics.precision_score(y_test, y_pred=pred , pos_label="1"))
print("recall ::", metrics.recall_score(y_test, y_pred=pred , pos_label="1"))
filename = 'SVM_model.sav'
joblib.dump(cls, filename)
# #to load model
# loaded_model = joblib.load(filename)
# result = loaded_model.score(x_test, y_test)
# print(result)
```
