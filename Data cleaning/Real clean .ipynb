```python
import pandas as pd 
import numpy as np 
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('CustomersDataset.csv')
df
```

```python
# step 1 : check for null values 
#________________________________________________
df.isnull().sum()
```

```python
# step 2 : check for wrong format values
# replace all wrong formatted data to No 
#_______________________________________________________
df = df.replace(['No phone service','No internet service'],'No')
df
```

```python
# step 3 
#________________________________________
# 1- convert data type of columns to suitable datatypes 
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce') # set all non numeric values to numeric 
df['TotalCharges'] = df['TotalCharges'].astype(float) # convert data type of TotalCharges column to float
```

```python
# 2 found null values then clean it 
df.dropna(inplace=True)
```

```python
# step 5 categoral data
#_________________________________________________________________________________
df.Contract = pd.Categorical(df.Contract , ['Month-to-month','One year','Two year'],ordered=True)
df.Contract = df.Contract.cat.codes

df.gender = pd.Categorical(df.gender , ['Male','Female'],ordered=True)
df.gender = df.gender.cat.codes

df.PaymentMethod = pd.Categorical(df.PaymentMethod , ['Electronic check','Mailed check','Bank transfer (automatic)','Credit card (automatic)'] ,ordered=True)
df.PaymentMethod = df.PaymentMethod.cat.codes

df.InternetService = pd.Categorical(df.InternetService , ['DSL','Fiber optic', 'No'] ,ordered=True)
df.InternetService = df.InternetService.cat.codes
```
